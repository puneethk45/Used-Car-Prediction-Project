# Removing Outliers

```python
for col in ['km_driven','selling_price','year','mileage','max_power','engine']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
   # print(IQR)
    lower_bound = Q1 - 1.5*IQR
   # print(lower_bound)
    upper_bound = Q3 + 1.5*IQR
   # print(upper_bound)
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
df.shape
```

```python
df['fuel'].unique()
```

```python
df['fuel'].replace(['Diesel','Petrol','LPG','CNG'],[0,1,2,3],inplace=True)
```

```python
df['seller_type'].replace(['Individual', 'Dealer', 'Trustmark Dealer'],[0,1,2],inplace=True)
df['transmission'].replace(['Manual', 'Automatic'],[0,1],inplace=True)
df['owner'].replace(['First Owner', 'Second Owner', 'Third Owner','Fourth & Above Owner'],[0,1,2,3],inplace=True)
```

#  Building the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn import metrics
```

```python
y=df['selling_price']
x=df.drop(columns=['selling_price','name','car_range'])
```

# Splitting into data into Train and Test set

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.15,random_state=0,shuffle=True)
```

# Linear Regression 

```python
linearRegression = LinearRegression()
linearRegression.fit(X_train,y_train)
predictions = linearRegression.predict(X_test)
print(r2_score(y_test,predictions))
```

```python
metrics.mean_absolute_error(y_test,predictions)
```

```python
metrics.mean_squared_error(y_test,predictions)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF=RandomForestRegressor()
RF.fit(X_train,y_train)
print('train accuracy: ',np.round(RF.score(X_train,y_train),3)*100,'% \n')
print('test accuracy: ',np.round(RF.score(X_test,y_test),3)*100,'% \n')
y_pred=RF.predict(X_test)
print(r2_score(y_pred,y_test))
```

# XGB regressor

```python
from xgboost import XGBRegressor
xg=XGBRegressor()
xg.fit(X_train,y_train)
y_pred=xg.predict(X_test)
print(r2_score(y_pred,y_test))
```

```python
print(xg.score(X_train,y_train))
```

```python
print(xg.score(X_test,y_test))
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dct=DecisionTreeRegressor()
dct.fit(X_train,y_train)
y_pred=dct.predict(X_test)
```

```python
r2_score(y_pred,y_test)
```
